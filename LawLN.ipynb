#How close is the sample mean to the population mean ?

The Law of Large Numbers tells us as the number of experiment reptitions increases, the average of the sample means better approximate the population mean

Any one sample or any one experiment is sensitive to sampling variability, noise, and other sources of non systemic variation

This means that one sample or one experiment is unlikely to provide a good estimate of true population mean



```python
#We want to generate a population of random numbers

#simulation parameters go here

pop_size = 325000
#print(3.25e5)

sample_size = 500
number_of_samples = 500
```

```python
#generate the population of numbers

#1/f distribution could use any d
#istribution because the sample mean
#even when the population is not
population = 1/np.logspace(np.log10(.001), np.log10(6),pop_size)

true_mean = np.mean(population)

#plot ever 1000 point
skip = int(1e3)
plt.plot(population[::skip], 'o')
plt.xlabel('Sample')
plt.ylabel('Data Vale')
plt.show()
```

```python
#shuffle the data so that they are not in any meaning temporal order
#random.suffle(population)

#we do not to set a variable equal to it
np.random.shuffle(population)

plt.plot(population[::skip], 'o')
plt.xlabel('Sample')
plt.ylabel('Data Vale')
plt.show()
```

Part Two Will involve Monte Carlo Sampling

We want to randomly select 50 of the 325000 data points and then compute their average
That will give me on point to plot and then we go through the population again and select another 50 samples and repeat a said amount of times, today will be 500

```python
#we need a numpy of zeros to store the sample means
sample_means = np.zeros(number_of_samples)

#np.random.choice() sample with replacement
#we also want to collec the mean of this
rand_sample =np.random.choice(population, sample_size)
#print(true_mean)
```

```python
#For loop to iterate over sample means and for every i collect the sample mean
# for i in range(number_of_samples):
#   sample_means[i] = np.mean(rand_sample)

for i in range(number_of_samples):
    sample_means[i] = np.random.choice(population, sample_size).mean()
# Calculate average for all sample mean
avg_sample_mean = sample_means.mean()
```

```python
#Plot the sample means
plt.plot(sample_means, 'ko', markerfacecolor='w', label='Sample Mean')
plt.plot([0, number_of_samples+10], [true_mean, true_mean], 'r', linewidth=5, label='True Mean')
plt.show()
```

The next step in the law of large numbers, is to consider the cummulative average of all the sample means.

```python
#cummulative average
cumave = np.zeros(number_of_samples)
for i in range(number_of_samples):
  cumave[i] = np.mean(sample_means[:i +1])
```

```python
plt.plot(cumave, 'ko', label='Cummulative Average')
plt.plot([0,number_of_samples], [true_mean,true_mean], 'r', linewidth=8, label='True Mean')
plt.xlabel('Sample Number')
plt.ylabel('Mean Value')
plt.show()
```
